# F-Tests

We first utilize an ANOVA test to determine the degree of the redundancy of the variables.

```r
anova(r_t_all)
```

We find $p$-values of $p(X_7) = 0.46, p(X_9) = 0.48, p(X_{10}) = 0.35$. All of these are higher than the $\alpha = 0.05$ threshold. Consequently we utilize a partial F-test to determine whether the partial model is sufficient.

```r
r_t_14568 <- lm(Y ~ X1 + X4 + X5 + X6, data = df_2)
summary(r_t_14568)
anova(r_t_14568, r_t_all)
```

The partial F-test has a $p$-value higher than our threshold-of-significance. Consequently we accept the null hypothesis and conclude $\beta_2 = \beta_3 = \beta_7 = \beta_9 = \beta_{10} = 0$. These corresponds to variables consumer expenditure, real wage growth, producer inflation rate, import price index, and export price index. We are left with real GDP growth, business and manufacturing labour productivity, inflation rate, and unemployment. We perform another ANOVA to consider the low t-test value in $X_6$, manufacturing labour productivity.

```r
anova(r_t_14568)
r_t_1458 <- lm(Y ~ X1 + X4 + X5 + X8, data = df_2)
summary(r_t_1458)
anova(r_t_1458, r_t_14568)
```

```r
anova(r_t_1458)
r_t_148 <- lm(Y ~ X1 + X4 + X8, data = df_2)
summary(r_t_148)
anova(r_t_148, r_t_1458)
```

Our ANOVA produces a high p-value for accepting the null hypothesis $\beta_6 = 0$. Therefore we remove manufacturing labour productivity and arrive at the real GDP growth, business labour productivity, inflation rate, and employment being the statistically-significant variables of our model.
