# Approval Rates

```python
ApprovalRates = pd.read_csv("STA302_FinalProject_RawData/STA302_FinalProject_ApprovalRates.csv", header=1)
ApprovalRates = pd.DataFrame(ApprovalRates)
ApprovalRates.drop(ApprovalRates.columns[[i for i in range(8, 25)]], axis=1, inplace=True)
ApprovalRates.drop(columns=["Geography", "Demographic", "Demographic Value", "Disapprove", "(DK/Refused)", "N Size"], inplace=True)
ApprovalRates.dropna(axis=0, inplace=True)
ApprovalRates.drop([i for i in range(5)], inplace=True)
ApprovalRates["Approve"] = ApprovalRates["Approve"].str.rstrip("%").astype("float") / 100
ApprovalRates.set_index("Time", inplace=True)
ApprovalRates = ApprovalRates.iloc[::-1]
ApprovalRates
```

# GDP Growth Rate

```python
GDPGrowthRates = pd.read_csv("STA302_FinalProject_RawData/STA302_FinalProject_RealGDPGrowthRates.csv", names=["Time", "GDP Growth Rate"])
GDPGrowthRates.drop(index=0, inplace=True)
GDPGrowthRates.drop([i for i in range(1, 209)], inplace=True)
GDPGrowthRates
```

```python
GDPGrowthRatesMonthly = pd.DataFrame(columns=["Time", "GDP Growth Rate"])
for i in range(len(GDPGrowthRates)):
    ymd = GDPGrowthRates.iloc[i, GDPGrowthRates.columns.get_loc("Time")].split("-")
    if ymd[1] == "01":
        months = ["Oct", "Nov", "Dec"]
        ymd[0] = str(int(ymd[0]) - 1)
    elif ymd[1] == "04":
        months = ["Jan", "Feb", "Mar"]
    elif ymd[1] == "07":
        months = ["Apr", "May", "Jun"]
    else:
        months = ["Jul", "Aug", "Sep"]
    for j in range(len(months)):
        GDPGrowthRatesMonthly.loc[len(GDPGrowthRatesMonthly) + j] = [f"{months[j]}/{ymd[0]}", 
                                                                     f"{GDPGrowthRates.iloc[i, GDPGrowthRates.columns.get_loc('GDP Growth Rate')]}"]
GDPGrowthRatesMonthly.set_index("Time", inplace=True)
GDPGrowthRatesMonthly
```

# Consumer Expenditure

```python
ConsumerExpenditure = pd.read_csv("STA302_FinalProject_RawData/STA302_FinalProject_ConsumerExpenditureRates.csv", names=["Time", "Consumer Expenditure"])
ConsumerExpenditure.drop(0, inplace=True)
ConsumerExpenditureMonthly = pd.DataFrame(columns=["Time", "Consumer Expenditure"])
months = {1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun", 7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"}
for i in range(len(ConsumerExpenditure)):
    ymd = ConsumerExpenditure.iloc[i, ConsumerExpenditure.columns.get_loc("Time")].split("-")
    ConsumerExpenditureMonthly.loc[len(ConsumerExpenditureMonthly)] = [f"{months[int(ymd[1])]}/{ymd[0]}", f"{ConsumerExpenditure.iloc[i, ConsumerExpenditure.columns.get_loc('Consumer Expenditure')]}"]
ConsumerExpenditureMonthly.drop([i for i in range(479)], inplace=True)
ConsumerExpenditureMonthly.set_index("Time", inplace=True)
ConsumerExpenditureMonthly
```

# Real Wage Growth

```python
RealWageGrowth = pd.read_csv("STA302_FinalProject_RawData/STA302_FinalProject_RealWageGrowth.csv")
RealWageGrowth.drop(25, inplace=True)
RealWageGrowth
```

```python
RealWageGrowthMonthly = pd.DataFrame(columns=["Time", "Real Wage Growth"])
for i in range(len(RealWageGrowth)):
    for j in range(1, len(RealWageGrowth.columns)):
        if j == 1:
            months = ["Jan", "Feb", "Mar"]
        elif j == 2:
            months = ["Apr", "May", "Jun"]
        elif j == 3:
            months = ["Jul", "Aug", "Sep"]
        else:
            months = ["Oct", "Nov", "Dec"]
        for m in months:
            RealWageGrowthMonthly.loc[len(RealWageGrowthMonthly)] = [f"{m}/{int(RealWageGrowth.iloc[i, RealWageGrowth.columns.get_loc('Year')])}", 
                                                                     f"{RealWageGrowth.iloc[i, RealWageGrowth.columns.get_loc(RealWageGrowth.columns[j])]}"]
RealWageGrowthMonthly.set_index("Time", inplace=True)
RealWageGrowthMonthly
```
